```python
import re

def extract_channel(notes):
    if pd.isna(notes):
        return 'unknown'
    
    found = re.search(r"Reached via\s+([^.;]+)", notes)
    if found:
        return found.group(1).strip().lower()
    
    return 'unknown'
```

```python
combined_all['first_contact_channel'] = combined_all["notes"].apply(extract_channel)
```

```python
combined_all.head()
```

Lead time in number of days

```python
combined_all['registration_lead_days'] = (combined_all['scheduled_date_time'] - combined_all['reg_timestamp']).dt.days
```

```python
combined_all.head()
```

Day of the week

```python
combined_all['event_day_of_week'] = combined_all['scheduled_date_time'].dt.day_name()
```

```python
combined_all.head()
```

```python
combined_all_final = pd.merge(combined_all,channels_df,on="event_id", how="inner")
```

Writing to db

```python
from sqlalchemy import create_engine

DBDIR = "C:\\Users\\VishwasKSingh\\Workspace\\temp\\my_app_practice\\dbs"
engine = create_engine(f"sqlite:///{DBDIR}/events.db")

with engine.begin() as conn:
    combined_all_final.to_sql(
        "events",
        con=conn,
        if_exists="replace",
        index=False
        
    )



```
